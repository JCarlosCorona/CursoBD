```python
import numpy as np
def main():
    """Funci√≥n principal del script"""
    # a X = b 3x3
    
    # Matriz de coeficientes
    a = np.array(
        [
            [2, 1, -3],
            [5, -4, 1],
            [1, -1, -4]
        ]
    )
    # Vector de resultados
    b = np.array([8, -19, 4])

    # Resolver el sistema de ecuaciones
    x = np.linalg.solve(a, b)
    
    print(x)
```
